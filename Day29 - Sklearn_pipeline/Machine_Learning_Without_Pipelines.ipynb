```python
# required data preprocessing libraries

import pandas as pd
import numpy as np
```

```python
da = pd.read_csv('train.csv')                     #infamous -titanic data
```

```python
da.head()
```

```python
da.drop(columns =['Name' , 'Ticket' , 'Cabin' , 'PassengerId'] , inplace = True)               #inplace = True eliminates the need of modifying means otherwise we need to assign the changes
#  to a new variable like new_da = .....
```

```python
da.head()
```

STEP - I

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(da.drop(columns = ['Survived']) , da['Survived'] , random_state = 42 , test_size = 0.2) #20%891 in testing - 178

# survived will be in o/p
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
da.isnull().sum()                      #check the missing value in all columns  - found the Embarked and age
```

```python
from sklearn.impute import SimpleImputer                   # to handle a missing data in a dataset

si_Age = SimpleImputer()
si_Embarked = SimpleImputer(strategy = 'most_frequent')                              # strategy means the method used to fill (impute) missing values.
# as here the Embarked is string we have taken mode otherwise by default it we will be taken as the mean not suitable for naN value to be filled out so used the mode and here it is S
# so msiising value get filled with the S(i.e 2values)

x_train_age_imp = si_Age.fit_transform(x_train[['Age']])
x_train_Embarked_imp = si_Embarked.fit_transform(x_train[['Embarked']])

# only transform
x_test_age_imp = si_Age.transform(x_test[['Age']])
x_test_Embarked_imp = si_Embarked.transform(x_test[['Embarked']])
```

```python
x_train_age_imp
x_train_Embarked_imp
```

```python
x_test_Embarked_imp
```

```python
da['Embarked'].unique()
```

STEP - II

```python
from sklearn.preprocessing import OneHotEncoder              #now for the categorical datasets -_-

# so for colums - Sex & Embarked
ohe_Sex = OneHotEncoder(sparse_output=False, handle_unknown= 'ignore')              # By default, OneHotEncoder returns a sparse matrix to save memory  but sparse=False,
# it returns a regular NumPy array (dense matrix)

ohe_Embarked = OneHotEncoder(sparse_output=False , handle_unknown = 'ignore')

# handle_unknown='ignore' This handles the case when, during prediction or transformation, the data has a category that the encoder didn’t see during training (fit)
# If don’t set this, it will give an error for unknown categories.  , like in training set it might not see that particular variabel not see it in the testset
# With handle_unknown='ignore', the encoder skips that unknown value and fills it with all zeros .
```

```python
x_train_Sex = ohe_Sex.fit_transform(x_train[['Sex']])
x_train_Embarked = ohe_Embarked.fit_transform(x_train_Embarked_imp)

x_test_Sex = ohe_Sex.transform(x_test[['Sex']])
x_test_Embarked = ohe_Embarked.transform(x_test_Embarked_imp)
```

```python
x_train_Sex   # two - male and female
```

```python
x_train_Embarked       # three - S C Q
```

```python
# now three arrays has been created for sex embarked and age

x_train_rem = x_train.drop(columns = ['Sex' , 'Embarked' , 'Age'])
x_test_rem = x_test.drop(columns = ['Sex' , 'Embarked' , 'Age'])
# this is for the rem cols means ie.4cols
```

```python
x_train_rem
```

```python
x_train_trans = np.concatenate(  (x_train_rem , x_train_Sex , x_train_Embarked , x_train_age_imp ) , axis = 1)
x_test_trans = np.concatenate(    (x_test_rem , x_test_Sex , x_test_Embarked  , x_test_age_imp ) , axis = 1)

#  axis=1 Join arrays column-wise i.e., side by side
```

```python
x_train_trans
```

```python
x_test_trans.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train_rem , y_train)

# trains (fits) the decision tree model using training data
```

```python
y_pred =  clf.predict(x_test_rem)
# uses the trained model (clf) to make predictions on the test data.
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

```python
import pickle         #pickle helps save Python objects (like models) as files

pickle.dump(ohe_Sex ,open('ohe_sex.pkl' , 'wb'))

# Saves  one-hot encoder for 'Sex' (ohe_sex) into a file.
# 'ohe_sex.pkl': path and filename where it's saved.
#'wb': means write in binary mode (needed for pickle files)

pickle.dump(ohe_Embarked , open('ohe_embarked.pkl' , 'wb'))

pickle.dump(clf , open('clf.pkl' , 'wb'))
# This saves trained decision tree model (clf) into a file
```
