PEFORMING NECESSARY OPERATIONS ON THE MIXED_VAR_COLN'S

```python
# ok so when the column contains both the numerical and categorical data then
# so now extracting the numerical part                     coerce - if values cannot be conbvertd to number then store itasNan,downcast
df['number_numerical'] = pd.to_numeric(df['number']  , errors='coerce' , downcast='integer')
df
# to save memory by using int types8,16
```

```python
# now extracting the categorical part
# from number_numeric ifits null means its char so set it from from the orign col ,else its nan
df['number_categorical'] = np.where(df['number_numerical'].isnull() , df['number'] , np.nan)
df
```

```python
# so now cabin and ticket has mixed_values
df['Cabin'].unique()
```

```python
#thenumeric partmethodwould notworkhere as they were seperatedvalues
# but herevlaues are aplhaumeric
df['Cabin_categorical']=df['Cabin'].str[0]   #aschar isinfirstindexonly
```

```python
df['Cabin_numerical']=df['Cabin'].str.extract('(\d+)')   #extract_thenumerical_part from string
df                           # \d+means oneormoredigit
```

```python
df['Cabin_categorical'].value_counts()   #how many_values_appears_inthe_column
```

```python
# now ticket contains no.of symbols,char,numbers aplhanumericstrings withsymbols

df['ticket_categorical']=df['Ticket'].str.extract('^([A-Za-z0-9./]+)')

# ^startofthestring [..]A character class thenallcharnumber + Match one or more of the allowed characters
```
