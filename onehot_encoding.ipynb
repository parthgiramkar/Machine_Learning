```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split( d.iloc[ : , 0 : 4 ] , d.iloc[: , -1]  , random_state = 1 , test_size = 0.2)
```

```python
x_train.head()                # 80% of 8128
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()      #by default produces a sparse matrix

# as sparse matrix is type of matrix where most valueare 0's only ,ex- if you have 1000 rows and 50 unique categories, your resulting matrix will be 1000×50, but each row will contain only one 1, and the rest will be 0.
# lot of waste space OneHotEncoder uses a sparse matrix by default to save memory and improve performance, especially when most values are zeros
# can change it to a normal (dense) array using sparse=False or .toarray()
```

```python
ohe.fit_transform(x_train[['fuel' , 'owner']] )       #9columns fuel-4 , owner-5
```

```python
ohe.fit_transform(x_train[['fuel' , 'owner']]).toarray
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel' , 'owner']]).toarray()   # as fit is only applied to the train only as so that data will not be loose to the train set , which is good practise
# here it's combines the fit& transform otherwise u can seperate it
x_test_new = ohe.transform(x_test[['fuel' , 'owner']]).toarray()
```

ONE HOTENCODING WITH TOP_CATEGORIES

```python
cnt = d.brand.value_counts()
cnt
```

```python
# selecting the freq/threshold that under it the brands will come under it
threshold = 100
```

```python
replace_ = cnt[cnt <= threshold].index
replace_
```

```python
pd.get_dummies(d['brand'].replace(replace_ , 'uncommon')).sample(4)      #here uncommon is the word that means that the rare brand with (less_freq<100) is merge with the uncommone label
# so all shown above in o/p comes under it
```
