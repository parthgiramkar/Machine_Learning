```python
import pandas as pd
```

```python
import numpy as np

df=pd.DataFrame( { 'age':[25,27,29,31,33,np.nan] , 'experience':[np.nan,3,5,7,9,11] ,
                'salary':[50,np.nan,110,140,170,200] ,'purchased' : [0,1,1,0,1,0] } )
```

```python
df
```

```python
x= df.drop(columns='purchased')         # remove from_the_feature_matrix o/p_column
y=df[['purchased']]
```

```python
x
```

```python
df.corr() #pandas_feature
```

```python
from sklearn.linear_model import BayesianRidge

by = BayesianRidge()
```

```python
from sklearn.experimental import enable_iterative_imputer  # The API and results of this estimator might change without any deprecation_cycle
# so set it
from sklearn.impute import IterativeImputer


imp = IterativeImputer(estimator=by , max_iter=10,random_state=40)
imp.fit_transform(x)
```
