```python
import pandas as pd
df = pd.read_csv('train (3).csv')
```

```python
df
```

```python
df = df[['Age','Survived','Fare']]
```

```python
x=df.drop(columns='Survived')
y=df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
x_train
```

```python
x_train.isnull().mean()*100
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer()
x_train_trf = si.fit_transform(x_train)
x_test_trf = si.transform(x_test)
```

```python
x_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(x_train_trf , y_train)

y_pred = clf.predict(x_test_trf)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred , y_test)
```

Missing_Indicator Technique to improve the accuracy

```python
from sklearn.impute import MissingIndicator
mi = MissingIndicator()

mi.fit(x_train)
```

```python
mi.features_ #tells that_firstcolumn_hasmissing_columns
```

```python
x_train_missing = mi.transform(x_train)
```

```python
x_train_missing.shape   # made byapplingtheage
```

```python
x_test_missing = mi.transform(x_test)
x_test_missing.shape
```

```python
x_train["age_NA"] = x_train_missing
x_train       #for no value then true
```

```python
x_test["age_NA"] = x_test_missing
x_test
```

```python
si = SimpleImputer()

x_train_trf2 = si.fit_transform(x_train)
x_test_trf2 = si.transform(x_test)
```

```python
x_train_trf2.shape
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(x_train_trf2 , y_train)

y_pred = clf.predict(x_test_trf2)
```

```python
# so the accuracy_increased by2% due to missing_indicator technique
from sklearn.metrics import accuracy_score
accuracy_score(y_pred , y_test)
```
