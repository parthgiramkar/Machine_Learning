```python
dt['date'] = pd.to_datetime(dt['date'])
dt.info()
# the object i.e string has been convertedto datetotime
```

EXTRACT Year/mon/day

```python
dt['date_year'] = dt['date'].dt.year
dt['date_month'] = dt['date'].dt.month
dt['date_day']  = dt['date'].dt.day

dt['date_month_name'] = dt['date'].dt.month_name()
dt.sample(8)
```

```python
dt['date_week']  = dt['date'].dt.day_of_week
dt['date_day_name']=dt['date'].dt.day_name()
dt.sample(8)
```

```python
 # the particular was weekend or not
import numpy as np

dt['date_isweekend'] = dt['date_day_name'].isin(['Saturday', 'Sunday'])
dt = dt.drop(columns=['product_id','city_id','orders'])
dt.sample(8)
```

```python
# np.where asign values accroding to the condn;s
dt['date_isweekend'] = np.where(dt['date_day_name'].isin(['Saturday', 'Sunday']) ,1,0)
dt['date_week'] = dt['date'].dt.isocalendar().week
dt.sample(8)
```
