```python
from sklearn.model_selection import train_test_split           #for testing 30% and then 70% for training
x_train , x_test , y_train , y_test = train_test_split( df.iloc[:,0:2] , df.iloc[ : , -1] , test_size = 0.3 , random_state = 1)
```

ORDINALENCODER CLASS OF SCIKIT-LEARN

```python
from sklearn.preprocessing import OrdinalEncoder
#created object of class ordinalencoder and in that taken two colmn i.e review and education(in formoflist)
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']] )

#here categories is the parameter that is used to define the order_as ordinal converts it into numbers
```

```python
# we do fit on the train and transform on the train and test
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```
