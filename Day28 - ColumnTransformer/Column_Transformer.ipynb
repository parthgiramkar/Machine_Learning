```python
# check if any value is missing or not
df.isnull().sum()
```

```python
# before tranformation train_test_split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['has_covid']) , df['has_covid'] , test_size=0.2 , random_state=1)
```

```python
x_train             # 80% of 100 - 80 entries
```

```python
# SimpleImputer is applied for the fever column , as it's used for handling missing values
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])
x_test_fever = si.transform(x_test[['fever']])
x_train_fever.shape
```

```python
# OneHotEnconding for gender and city as it converts nominal data into the vectors form
# for gender we have - 2cols i.e male&female means only one will be taken & form  city we have 4 so 3 will be taken
# why ? becasue of the multicollinearity , so in total there are 4 cols

ohe = OneHotEncoder(drop= 'first' , sparse_output =False)
# drop the one_col from each category and sparse_output returns a regular NumPy array instead of a sparse matrix


x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])
x_test_gender_city  = ohe.transform(x_test[['gender','city']])

print(x_train_gender_city .shape)
x_train_gender_city
```

```python
# so gender&city are nominal_categorical data , cough can be compared so ordinal
oe = OrdinalEncoder(categories=[['Mild' , 'Strong']])

x_train_cough = oe.fit_transform(x_train[['cough']])
x_test_cough = oe.transform(x_test[['cough']])

x_train_cough.shape
```

```python
# now the age as its remanining
# if not used .values() it willreturn resultin pandas df not numpy array so ,now it hasconv it_into numpy array
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values

print(x_train_age)
x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_cough , x_train_fever ,x_train_gender_city , x_train_age),axis=1)
x_test_transformed = np.concatenate((x_test_cough , x_test_fever ,x_test_gender_city , x_test_age),axis=1)

# To horizontally join columns — side by side.

print(x_train_transformed)
x_train_transformed.shape
```

WORKING WITH COLUMN_TRANSFORMER

```python
# with columntransformer we can do all inonce ,noneed for seperate applying transformation
# apply different preprocessing steps to different columns of your dataset — all in one step.
from sklearn.compose import ColumnTransformer
transform = ColumnTransformer(transformers = [('tf1' , SimpleImputer() , ['fever'] ) , ( 'tf2' ,OneHotEncoder(drop= 'first' , sparse_output =False) , ['gender','city'] ) , ( 'tf3' ,OrdinalEncoder(categories=[['Mild' , 'Strong']] ) , ['cough'] ) ] , remainder = 'passthrough')


# remainder='passthrough' - means - keep the rest of the columns as they are, don't drop them
# so , now passing reqd. tuples inside the transfomers list with the syntax - Tuple format	('name', transformer, columns)
```
