```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
df
fd = df[['Age','Pclass','SibSp','Parch','Survived']]
fd
```

```python

fd.info()
```

```python
# then the drop the Age column missing values
fd.dropna(inplace=True)
```

```python
fd
```

```python
x=fd.drop(columns='Survived')
y=fd.iloc[:, -1]
```

```python
x.tail()
```

 Use cross_val_score when dataset is not huge

```python
# Splits the data into 20 parts (folds)
# Runs the model 20 times, each time using a different fold as test data ya but 1fold is thetest_set
# cross_val_score(..) - returns the array of 20accur_scores and np.mean() gives the average of all
# x, y is the data while the , LogisticRegression() is the model.

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

np.mean(  cross_val_score ( LogisticRegression(), x,y,scoring="accuracy",cv=20) )
```

APPLYING THE FEATURE CONSTRUCTION

```python
x['family_size'] = x['Parch'] + x['SibSp'] + 1
x
```

```python
def myhome(num) :
  if num == 1 :
    return 0          #mingle

  elif num >1 and num<=4 :
    return 1        #nuclear family

  else :
    return 2  #large family
```

```python
myhome(5)
```

```python
x['family_type'] = x['family_size'].apply(myhome)
x
```

```python
x.drop(columns=['SibSp',"Parch","family_size"])
```

```python
np.mean( cross_val_score ( LogisticRegression() , x,y,scoring="accuracy" , cv=20) )

```
